```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, norm, uniform
import seaborn as sns
```

# 2.

```python
arr_uniform = np.random.uniform(0, 1, 1000)
arr_normal_1 = np.random.normal(0, 1, 1000)
arr_normal_10 = np.random.normal(0, 10, 1000)
arr_beta = np.random.beta([1, 2, 3], [1, 2, 9], (1000, 3))
arr_two_dim_1 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10000)
arr_two_dim_2 = np.random.multivariate_normal([0, 0], [[2, 1], [1, 2]], 10000)
```

# 3.

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(11, 8))
bins = 20
ax1.hist(arr_uniform, bins=bins, density=True)
ax1.set_title('uniform distribution')
ax2.hist(arr_normal_1, bins=bins, density=True)
ax2.set_title('normal distribution (0, 1)')
ax3.hist(arr_normal_10, bins=bins, density=True)
ax3.set_title('normal distribution (0, 10)')
ax4.hist(arr_beta[:, 0], bins=bins, density=True)
ax4.set_title('beta distribution')
ax5.hist(arr_beta[:, 1], bins=bins, density=True)
ax5.set_title('two-dimensional normal distributions')
ax6.hist(arr_beta[:, 2], bins=bins, density=True)
ax6.set_title('two-dimensional normal distributions')
plt.show()
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(11, 8))
plt.suptitle('normalized histograms with plots of analytical formulas for distributions')
bins = 20

ax1.hist(arr_uniform, bins=bins, density=True)
x = np.linspace(0, 1, 1000)
ax1.plot(x, uniform.pdf(x, 0, 1))
ax1.set_title('uniform distribution')

ax2.hist(arr_normal_1, bins=bins, density=True)
x = np.linspace(-4, 4, 1000)
ax2.plot(x, norm.pdf(x, 0, 1))
ax2.set_title('normal distribution (0, 1)')

ax3.hist(arr_normal_10, bins=bins, density=True)
x = np.linspace(-30, 30, 1000)
ax3.plot(x, norm.pdf(x, 0, 10))
ax3.set_title('normal distribution (0, 10)')

ax4.hist(arr_beta[:, 0], bins=bins, density=True)
x = np.linspace(0, 1, 1000)
ax4.plot(x, beta.pdf(x, 1, 1))
ax4.set_title('beta distribution')

ax5.hist(arr_beta[:, 1], bins=bins, density=True)
x = np.linspace(0, 1, 1000)
ax5.plot(x, beta.pdf(x, 2, 2))
ax5.set_title('two-dimensional normal distributions')

ax6.hist(arr_beta[:, 2], bins=bins, density=True)
x = np.linspace(0, 1, 1000)
ax6.plot(x, beta.pdf(x, 3, 9))
ax6.set_title('two-dimensional normal distributions')

plt.show()
```

# 5.

```python
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
gamma = np.abs(np.random.normal(0, 1, 1000))

mu = alpha + beta * x
y = np.random.normal(mu, gamma, 1000)
```

# 6.

```python
plt.scatter(x, y)
x = np.linspace(-6, 6, 1000)
for i in range(1000):
    plt.plot(x, alpha[i] + beta[i] * x, 'r')
plt.grid()
plt.show()
```
